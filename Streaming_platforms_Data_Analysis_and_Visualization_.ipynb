## 2.1 Change IMDB rating to categories
3 different categories given the ratings:
- Good (7 to 10)
- Average (5 to 7)
- Bad (0 to 5)

```python
def imdb_rating(rating):
    if rating >= 7:
        return 'Good'
    elif rating >= 5:
        return 'Average'
    else:
        return 'Bad'

moviesDF['IMDb_Categories'] = moviesDF['IMDb'].apply(imdb_rating)
print(moviesDF[['IMDb', 'IMDb_Categories']].head())
```

## 2.2 Clean NULL values
Change `NA` values in the column `age` to `Not Rated`

```python
moviesDF['Age'] = moviesDF['Age'].fillna('Not Rated')
```

```python
plt.figure(figsize=(12, 6))

for category, color in zip(['Good', 'Average', 'Bad'], ['orange', 'blue', 'red']):
    subset = moviesDF[moviesDF[category] == 1]
    plt.scatter(subset['Year'], subset['Runtime'], alpha=0.5, color=color, label=category)

plt.title('Runtime vs Release Year with IMDb Categories')
plt.xlabel('Release Year')
plt.ylabel('Runtime (minutes)')
plt.legend()
plt.grid(True)
plt.show()
```

The scatter plot categorizes movies by IMDb rating categories (Good, Average, Bad) against their runtime and release year. Most films, regardless of rating, fall into a consistent runtime band. There's no clear trend linking runtime to rating quality.
However, it can be observed that over time, the number of recent movies rated bad is significantly higher than that of older movies

### 3.9a Analyse the frequency of the words then set a threshold for visualization

```python
!pip install nltk
```

```python
import nltk
from nltk.probability import FreqDist
nltk.download('punkt')

all_genres = ' '.join(moviesDF['Genres'].dropna().str.replace(', ', ' '))
tokens = nltk.word_tokenize(all_genres)
freq_dist = FreqDist(tokens)

frequent_threshold = freq_dist.most_common(1)[0][1] * 0.5
rare_threshold = 5

filtered_freq_dist = {word: freq for word, freq in freq_dist.items() if freq < frequent_threshold and freq > rare_threshold}

wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(filtered_freq_dist)
plt.figure(figsize=(15, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## RQ1 (Multiples subplots): How has the availability of movies on different streaming platforms changed over the years?

```python
import matplotlib.pyplot as plt
moviesDF[['Netflix', 'Prime Video', 'Hulu', 'Disney+']] = moviesDF[['Netflix', 'Prime Video', 'Hulu', 'Disney+']].fillna(0)

availability_per_year = moviesDF.groupby('Year')[['Netflix', 'Prime Video', 'Hulu', 'Disney+']].sum().reset_index()

fig, axes = plt.subplots(2, 2, figsize=(14, 10), sharex=True, sharey=True)
fig.suptitle('Availability of Movies on Different Streaming Platforms Over Time')

streaming_platforms = ['Netflix', 'Prime Video', 'Hulu', 'Disney+']
for i, platform in enumerate(streaming_platforms):
    ax = axes[i//2, i%2]
    ax.plot(availability_per_year['Year'], availability_per_year[platform], label=platform)
    ax.set_title(platform)
    ax.set_xlabel('Year')
    ax.set_ylabel('Count of Movies')
    ax.grid(True)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

The availability of movies on different streaming platforms has shown significant changes over the years as illustrated in the graphs:

*   Netflix and Prime Video display a similar trend with a substantial increase in the number of movies available in recent years. This uptrend suggests aggressive content acquisition strategies or an increase in production.
*   Hulu shows a gradual increase over time with less pronounced growth than Netflix and Prime Video.
*   Disney+ has a relatively flat trend for most of the timeline, with a slight increase visible in the most recent years, indicating a more conservative addition of movies to their platform or a newer entry into the streaming market.




## RQ2 (Pivot table): Are there significant differences in the average Runtime of movies on different streaming platforms?

```python
platforms = ['Netflix', 'Prime Video', 'Hulu', 'Disney+']
runtime_data = {
    'Platform': platforms,
    'Average Runtime': [moviesDF[moviesDF[platform] == 1]['Runtime'].mean() for platform in platforms]
}

runtime_df = pd.DataFrame(runtime_data)

plt.figure(figsize=(10, 6))
plt.bar(runtime_df['Platform'], runtime_df['Average Runtime'], color=['red', 'blue', 'green', 'orange'])
plt.title('Average Movie Runtime on Different Streaming Platforms')
plt.ylabel('Average Runtime (minutes)')
plt.xlabel('Streaming Platform')
plt.xticks(rotation=45)
plt.show()
```

Based on the bar chart, there do not appear to be significant differences in the average runtime of movies across the streaming platforms Netflix, Prime Video, Hulu, and Disney+. All platforms show average movie runtimes that are quite close to each other, typically within the standard feature-length film window. This suggests that despite the diversity of content across platforms, the average length of a movie is consistent, likely adhering to industry standards for feature-length productions.







## RQ3: Which countries produce movies with high IMDb ratings？

```python
filtered_movies = moviesDF.dropna(subset=['IMDb', 'Country'])

countries_expanded = filtered_movies.assign(Country=filtered_movies['Country'].str.split(',')).explode('Country')

country_ratings = countries_expanded.groupby('Country')['IMDb'].mean().sort_values(ascending=False)

top_countries = country_ratings.head(10)

top_countries.plot(kind='bar', figsize=(12, 6))
plt.title('Top 10 Countries with the Highest Average IMDb Ratings')
plt.ylabel('Average IMDb Rating')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.show()
```

Bar chart showing the top 10 countries with the highest average IMDb movie ratings. Ratings in these countries are particularly high, averaging above 7, indicating a high quality film. However, smaller countries or countries with lower film industry production may have fewer films rated, which may result in higher average ratings than countries with greater film production.

For example, countries like Moldova, the Republic of North Macedonia, and Montenegro are not generally known for large-scale film production, so the best films from these countries may be rated on IMDb. In contrast, film powerhouses such as the United States or India may not be rated on IMDb due to the sheer volume of films they produce and the wider range of film ratings.

This also shows that a movie's IMDb rating is affected by many factors, and it is not ideal to just rely on country and average rating to select the highest movie
